```python
from collections import Counter
print("Question 1")
my_list = [2, 3, 6]
print(my_list)
result = 1
for item in my_list:
    result = result * item
print("multiplication of all elements : ", result)
print("---------------------------------")
print("Question 2")
numbers = [(2, 5), (1, 2), (4, 4), (2, 3), (2, 1)]
print("Before:", numbers)
numbers.sort()
print("After:", numbers)
print("---------------------------------")
print("Question 3")
d1 = {'a': 100, 'b': 200, 'c': 300}
d2 = {'a': 300, 'b': 200, 'd': 400}
d3 = Counter(d1) + Counter(d2)
print(d3)
print("---------------------------------")
print("Question 4")
n = int(input())
d = dict()
for i in range(1, n+1):
    d[i] = i*i
print(d)
print("---------------------------------")
print("Question 5")
list = [('item1', '12.20'), ('item2', '15.10'), ('item3', '24.5')]
print("before:", list)


def Sort(tup):
    return (sorted(tup, key=lambda x: float(x[1]), reverse=True))


print("After:", Sort(list))
print("---------------------------------")
print("Question 6")
num_set = set([0, 1, 2, 3, 4])
print(num_set)
for iteration in num_set:
    print(iteration)
num_set.add(5)
num_set.remove(0)
print(num_set)
for iteration in num_set:
    print(iteration)
```
