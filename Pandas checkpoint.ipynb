```python
import pandas as pd
import numpy as np

def sep():
    print("-" * 50)

exam_data = {
    'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
    'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
    'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']
}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

exam_data_pd = pd.DataFrame(exam_data, index=labels)

print(exam_data_pd.head(3))
sep()

exam_data_pd = exam_data_pd.dropna()
print(exam_data_pd)
sep()

name_score = exam_data_pd[['name', 'score']]
print(name_score)
sep()

new = {'name': 'Suresh', 'score': 15.5, 'attempts': 1, 'qualify': 'yes'}
new = pd.DataFrame(new, index=['k'])
exam_data_pd = pd.concat([exam_data_pd, new])
print(exam_data_pd)
sep()

del exam_data_pd['attempts']
print(exam_data_pd)
sep()

def result(score):
    if score > 10:
        return 1
    else:
        return 0

exam_data_pd['Success'] = exam_data_pd['score'].apply(result)
print(exam_data_pd)
sep()

exam_data_pd.to_csv('my_data.csv')
```
