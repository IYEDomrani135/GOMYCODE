```python
from tkinter import *
import random
root = Tk()
root.title("Rock Paper Scissors")
root.geometry("500x200")
root.resizable(width=False, height=False)
root.configure(bg="black")

L1 = Label(text="Rock Paper Scissors", bg="black", fg="white")
L1.place(x=210, y=0)

L2 = Label(text="Choose between rock, paper, or scissors.", bg="black", fg="white")
L2.place(x=2, y=30)

user_move = Entry(root)
user_move.place(x= 5, y= 60)

comp_score = 0
user_score = 0

def display_score():
    global comp_score, user_score
    Score_display = Label(text=f"User's score: {user_score}", width=12, bg="black", fg="green")
    Score_display.place(x=2, y=110)
    Score_display2 = Label(text=f"Computer's score: {comp_score}", width=15, bg="black", fg="red")
    Score_display2.place(x=378, y=110)

def play():
    global comp_score, user_score

    user_pick = user_move.get()

    if user_pick == 'rock' or user_pick == 'paper' or user_pick == 'scissors':
        L3 = Label(text=f"Chose {user_pick}",width=10, bg="black", fg="green")
        L3.place(x=2, y=90)

        Moves = ["rock", "paper", "scissors"]
        comp_pick = random.choice(Moves)
        L3 = Label(text=f"Chose {comp_pick}", width=10 ,bg="black", fg="red")
        L3.place(x=410, y=90)

        if user_pick == "rock" and comp_pick == "rock":
            result = "Tie"
        elif user_pick == "rock" and comp_pick == "paper":
            result = "Lost"
            comp_score += 1
        elif user_pick == "rock" and comp_pick == "scissors":
            result = "Win"
            user_score += 1
        display_score()
        
        if user_pick == "paper" and comp_pick == "rock":
            result = "Win"
            user_score += 1
        elif user_pick == "paper" and comp_pick == "paper":
            result = "Tie"
        elif user_pick == "paper" and comp_pick == "scissors":
            result = "Lost"
            comp_score += 1

        if user_pick == "scissors" and comp_pick == "rock":
            result = "Lost"
            comp_score += 1
        elif user_pick == "scissors" and comp_pick == "paper":
            result = "Win"
            user_score += 1
        elif user_pick == "scissors" and comp_pick == "scissors":
            result = "Tie"
        if result == "Tie":
            result_color = "orange"
        elif result == "Win":
            result_color = "green"
        elif result == "Lost":
            result_color = "red"
        else:
            pass
        display_score()
    L4 = Label(text=f"{result}",width=10, bg="black", fg=f"{result_color}")
    L4.place(x=220, y=90)

def reset():
    global comp_score, user_score
    comp_score = 0
    user_score = 0
    display_score()

Play = Button(text="Play", command= play)
Play.place(x=150, y=58, width=100, height= 25)

reset_btn = Button(text="Reset", command= reset)
reset_btn.place(x=10, y=170, width=100, height= 25)

def Exit():
   root.destroy()

exit_btn = Button(text="Exit", command= Exit)
exit_btn.place(x=390, y=170, width=100, height= 25)

root.mainloop()
```
