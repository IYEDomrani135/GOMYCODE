```python
def sep():
    print('-'*50)
#1
binary_search1 = ([1,2,3,5,8], 6) # False
binary_search2 = ([1,2,3,5,8], 5) # True
def binary_search(arr):
    array_to_search, target = arr
    low, high = 0, len(array_to_search) - 1
    while low <= high:
        mid = (low + high) // 2
        if array_to_search[mid] == target:
            return True
        elif array_to_search[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return False

print(f"{binary_search(binary_search1)}")
print(f"{binary_search(binary_search2)}")
sep()
#2
def power(a, b):
    result = a**b
    return result
inp_values_1 = int(input("Number: "))
inp_values_2 = int(input(f"{inp_values_1} to the power: "))
print(f"{inp_values_1} to the power of {inp_values_2} is equal to: {power(inp_values_1, inp_values_2)}")
sep()
#3
Sample_Data = [29,13,22,37,52,49,46,71,56] # Expected result: [13, 22, 29, 37, 42, 46, 49, 56, 71]
def bubble(Sample_Data):
    for passnum in range(len(Sample_Data)-1,0,-1):
        for i in range(passnum):
            if Sample_Data[i]>Sample_Data[i+1]:
                temp = Sample_Data[i]
                Sample_Data[i] = Sample_Data[i+1]
                Sample_Data[i+1] = temp
    return Sample_Data
print(f"Sorted list: {bubble(Sample_Data)}")
sep()
#4
def merge_sort(arr):
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])
    return merge(left, right)

def merge(left, right):
    result = []
    left_index, right_index = 0, 0

    while left_index < len(left) and right_index < len(right):
        if left[left_index] < right[right_index]:
            result.append(left[left_index])
            left_index += 1
        else:
            result.append(right[right_index])
            right_index += 1

    result.extend(left[left_index:])
    result.extend(right[right_index:])
    return result

sample_data = [29, 13, 22, 37, 52, 49, 46, 71, 56] # Output: [13, 22, 29, 37, 46, 49, 52, 56, 71]
sorted_data = merge_sort(sample_data)
print(sorted_data)  
sep()
#5
sample_data = [29, 13, 22, 37, 52, 49, 46, 71, 56]
def quick_sort(sample_data):
    if len(sample_data) <= 1:
        return sample_data
    
    pivot = sample_data[len(sample_data) // 2]
    left = [x for x in sample_data if x < pivot]
    middle = [x for x in sample_data if x == pivot]
    right = [x for x in sample_data if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

sorted_data = quick_sort(sample_data)
print(sorted_data)
```
