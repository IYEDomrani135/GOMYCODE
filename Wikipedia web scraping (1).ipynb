```python
import requests
from bs4 import BeautifulSoup
url = 'https://en.wikipedia.org/wiki/Python_(programming_language)'

def page(url):
    article_paragraph = {}
    article_links = []
    src = requests.get(url)
    soup = BeautifulSoup(src.content, 'html.parser')

    def title(soup):
        return soup.find('span',{'class':'mw-page-title-main'}).text.strip()
    title_page = title(soup)

    def paragraph(soup):
        for element in soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6', 'p']):
            if element.name.startswith('h'):
                current_heading = element.get_text()
                article_paragraph[current_heading] = []
            elif element.name == 'p' and current_heading:
                article_paragraph[current_heading].append(element.get_text())
    paragraph(soup)

    def links(soup):
        for link in soup.find_all('a', href=True):
            article_links.append(link['href'])
    links(soup)
    
    return {
        'title': title_page,
        'article_paragraph': article_paragraph,
        'redirect_links': article_links
    }

result = page(url)
print(result)
```
